### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
```

```python
print(f"Faturamento Loja 1: R$ {faturamento_loja:,.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:,.2f}")
```

# 2. Vendas por Categoria


```python
categorias_loja = loja['Categoria do Produto'].value_counts()
categorias_loja2 = loja2['Categoria do Produto'].value_counts()
categorias_loja3 = loja3['Categoria do Produto'].value_counts()
categorias_loja4 = loja4['Categoria do Produto'].value_counts()
```

```python
print("Produtos vendidos por categoria - Loja 1:")
print(categorias_loja, "\n")

print("Produtos vendidos por categoria - Loja 2:")
print(categorias_loja2, "\n")

print("Produtos vendidos por categoria - Loja 3:")
print(categorias_loja3, "\n")

print("Produtos vendidos por categoria - Loja 4:")
print(categorias_loja4, "\n")
```

```python
mais_vendida_loja1 = loja['Categoria do Produto'].value_counts().idxmax()
mais_vendida_loja2 = loja2['Categoria do Produto'].value_counts().idxmax()
mais_vendida_loja3 = loja3['Categoria do Produto'].value_counts().idxmax()
mais_vendida_loja4 = loja4['Categoria do Produto'].value_counts().idxmax()
```

```python
print(f"Categoria mais vendida na Loja 1: {mais_vendida_loja1}")
print(f"Categoria mais vendida na Loja 2: {mais_vendida_loja2}")
print(f"Categoria mais vendida na Loja 3: {mais_vendida_loja3}")
print(f"Categoria mais vendida na Loja 4: {mais_vendida_loja4}")
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()
```

```python
print(f"Média de avaliação - Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de avaliação - Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de avaliação - Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de avaliação - Loja 4: {media_avaliacao_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
def produto_destaque(df, nome_loja):
    contagem = df['Produto'].value_counts()
    mais_vendido = contagem.idxmax()
    menos_vendido = contagem.idxmin()
    print(f"{nome_loja} - Produto mais vendido: {mais_vendido} ({contagem.max()} vendas)")
    print(f"{nome_loja} - Produto menos vendido: {menos_vendido} ({contagem.min()} vendas)\n")
```

```python
produto_destaque(loja, "Loja 1")
produto_destaque(loja2, "Loja 2")
produto_destaque(loja3, "Loja 3")
produto_destaque(loja4, "Loja 4")
```

```python
def grafico_top_vendas(df, nome_loja):
    top10 = df['Produto'].value_counts().head(10)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=top10.values, y=top10.index, palette="viridis")
    plt.title(f"Top 10 Produtos Mais Vendidos - {nome_loja}")
    plt.xlabel("Quantidade Vendida")
    plt.ylabel("Produto")
    plt.tight_layout()
    plt.show()
```

```python
grafico_top_vendas(loja, "Loja 1")
grafico_top_vendas(loja2, "Loja 2")
grafico_top_vendas(loja3, "Loja 3")
grafico_top_vendas(loja4, "Loja 4")
```

#6. Gráficos

* **Faturamento Total por Loja**

```python
faturamentos = [
    loja['Preço'].sum(),
    loja2['Preço'].sum(),
    loja3['Preço'].sum(),
    loja4['Preço'].sum()
]

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 5))
sns.barplot(x=lojas, y=faturamentos, palette="Blues_d")
plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento (R$)")
plt.xlabel("Loja")
plt.tight_layout()
plt.show()
```

```python
todas_lojas = pd.concat([loja, loja2, loja3, loja4])

categorias_total = todas_lojas['Categoria do Produto'].value_counts()


plt.figure(figsize=(7, 7))
plt.pie(categorias_total, labels=categorias_total.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Distribuição das Categorias de Produto - Todas as Lojas")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
avaliacoes = [
    loja['Avaliação da compra'].mean(),
    loja2['Avaliação da compra'].mean(),
    loja3['Avaliação da compra'].mean(),
    loja4['Avaliação da compra'].mean()
]

plt.figure(figsize=(8, 5))
sns.barplot(x=lojas, y=avaliacoes, palette="Greens_d")
plt.title("Média das Avaliações dos Clientes por Loja")
plt.ylabel("Nota média")
plt.ylim(0, 5)
plt.xlabel("Loja")
plt.tight_layout()
plt.show()
```
